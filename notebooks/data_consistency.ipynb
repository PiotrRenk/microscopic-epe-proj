```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
data = pd.read_csv('../materials/processed_df_new.csv')
```

```python
def check_consistency(cols):
    for col1, col2, aggr_col in cols:
        max_ = data[[col1, col2]].max(axis=1)
        matches = max_ == data[aggr_col]

        if not matches.all():
            print(f"Inconsistency found in {aggr_col}")
```

```python
test_cols = [
    ('Bx ISUP Grade P', 'Bx ISUP Grade L', 'Bx ISUP Grade'),
    ('ISUP RP P', 'ISUP RP L', 'ISUP RP'),
    ('MRI EPE P', 'MRI EPE L', 'MRI EPE'),
    ('MRI SVI P', 'MRI SVI L', 'MRI SVI'),
    ('EPE P RP', 'EPE L RP', 'EPE RP'),
    ('SVI RP P', 'SVI RP L', 'SVI RP')
]
check_consistency(test_cols)
```

```python
ISUP_max = data[['ISUP RP P', 'ISUP RP L']].max(axis=1)
match_ = ISUP_max == data['ISUP RP']

total = len(data)
matches = match_.sum()
print(f"Total rows: {total}")
print(f"Matches: {matches}")
data.loc[match_ == False]
```
