# Outliers

```python
def count_outliers(df, numerical_cols):
    print(f"Number of outliers in:")
    for column in numerical_cols:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        count = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
        print(f"  - {column}: {count} ({count/len(df)*100:.2f}%)")

numerical_cols = ['wiek', 'PSA', 'PSAdensity', 'MRI vol', 'MRI SIZE']
count_outliers(df, numerical_cols)
```

Considerable number of outliers are present but they probably stem from extreme cases of illness so we don't remove them.
