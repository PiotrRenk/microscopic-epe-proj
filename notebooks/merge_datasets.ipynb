```python
def clean_first(first_data):
    # Rename column 'NEW IDENTIFIER' to 'ID'
    first_data = first_data.rename(columns={'NEW IDENTIFIER': 'ID'})
    # Rename column 'DRE (0-prawidłowe; 1-nieprawidłowe)' to 'DRE'
    first_data = first_data.rename(columns={'DRE (0-prawidłowe; 1-nieprawidłowe)': 'DRE'})
    # Drop column: 'data operacji'
    first_data = first_data.drop(columns=['data operacji'])
    # Rename column 'PSAdensity' to 'PSA density'
    first_data = first_data.rename(columns={'PSAdensity': 'PSA density'})
    # Drop column: 'lokalizacja'
    first_data = first_data.drop(columns=['lokalizacja'])
    # Rename column 'MRI EPE (naciek poza torebke)' to 'MRI EPE'
    first_data = first_data.rename(columns={'MRI EPE (naciek poza torebke)': 'MRI EPE'})
    # Rename column 'MRI SVI (pecherzyki)' to 'MRI SVI'
    first_data = first_data.rename(columns={'MRI SVI (pecherzyki)': 'MRI SVI'})
    # Rename column 'MRI SIZE INDEX LESION' to 'MRI SIZE'
    first_data = first_data.rename(columns={'MRI SIZE INDEX LESION': 'MRI SIZE'})
    # Rename column 'ilość\xa0+ wycinków P' to 'wycinki P'
    first_data = first_data.rename(columns={'ilość\xa0+ wycinków P': 'wycinki P'})
    # Rename column 'Ilość + wycinków L' to 'wycinki L'
    first_data = first_data.rename(columns={'Ilość + wycinków L': 'wycinki L'})
    # Rename column 'SM +' to 'SM+'
    first_data = first_data.rename(columns={'SM +': 'SM+'})
    # Rename column 'N +' to 'N+'
    first_data = first_data.rename(columns={'N +': 'N+'})
    # Rename column 'EPE L RP' to 'EPE RP L'
    first_data = first_data.rename(columns={'EPE L RP': 'EPE RP L'})
    # Rename column 'EPE P RP' to 'EPE RP P'
    first_data = first_data.rename(columns={'EPE P RP': 'EPE RP P'})
    # Drop column: 'Kolumna2'
    first_data = first_data.drop(columns=['Kolumna2'])
    # Drop column: 'Kolumna3'
    first_data = first_data.drop(columns=['Kolumna3'])
    # Drop column: 'Prostate vol '
    first_data = first_data.drop(columns=['Prostate vol '])
    # Rename column 'MRI SIZE' to 'MRI lesion size'
    first_data = first_data.rename(columns={'MRI SIZE': 'MRI lesion size'})
    return first_data

first_data_clean = clean_first(first_data.copy())
first_data_clean.head()
```

```python
def getCores(patient, cores_P_col='ilość\xa0+ wycinków P', cores_L_col='Ilość + wycinków L'):
    cores_P = patient[cores_P_col]
    if cores_P == 0 or pd.isna(cores_P):
        pos_P = 0
        neg_P = 0
    else:
        cores_P = cores_P.split('na')
        pos_P = int(cores_P[0])
        neg_P = int(cores_P[1]) - pos_P

    cores_L = patient[cores_L_col]
    if cores_L == 0 or pd.isna(cores_L):
        pos_L = 0
        neg_L = 0
    else:
        cores_L = cores_L.split('na')
        pos_L = int(cores_L[0])
        neg_L = int(cores_L[1]) - pos_L

    return pos_P, neg_P, pos_L, neg_L
```

```python
def merge_percent(row):
    percent_P = row['% P']
    percent_L = row['% L']
    if pd.isna(percent_P) and pd.isna(percent_L):
        return np.nan
    # elif pd.isna(percent_P):
    #     return percent_L
    # elif pd.isna(percent_L):
    #     return percent_P

    percent_P = row['% P'] if not pd.isna(row['% P']) else 0
    percent_L = row['% L'] if not pd.isna(row['% L']) else 0
    return np.mean((percent_P, percent_L))
```

```python
def first_cores(patient, cores_P_col='ilość\xa0+ wycinków P', cores_L_col='Ilość + wycinków L'):
    cores_P = patient[cores_P_col]
    if cores_P == 0 or pd.isna(cores_P):
        pos_P = 0
        neg_P = 0
    else:
        cores_P = cores_P.split('na')
        pos_P = int(cores_P[0])
        neg_P = int(cores_P[1]) - pos_P

    cores_L = patient[cores_L_col]
    if cores_L == 0 or pd.isna(cores_L):
        pos_L = 0
        neg_L = 0
    else:
        cores_L = cores_L.split('na')
        pos_L = int(cores_L[0])
        neg_L = int(cores_L[1]) - pos_L

    return pos_P, neg_P, pos_L, neg_L, pos_P + pos_L, neg_P + neg_L

def first_percent(row):
    percent_P = row['% P']
    percent_L = row['% L']
    if pd.isna(percent_P) and pd.isna(percent_L):
        return np.nan
    elif pd.isna(percent_P):
        return percent_L
    elif pd.isna(percent_L):
        return percent_P

    # percent_P = row['% P'] if not pd.isna(row['% P']) else 0
    # percent_L = row['% L'] if not pd.isna(row['% L']) else 0
    return np.mean((percent_P, percent_L))
```

```python
def process_first(first_data):
    first_data = first_data.drop(columns=['data operacji', 'lokalizacja', 'Kolumna2', 'Kolumna3', 'Prostate vol '])
    first_data = first_data.rename(columns={'NEW IDENTIFIER': 'ID',
                                            'DRE (0-prawidłowe; 1-nieprawidłowe)': 'DRE',
                                            'PSAdensity': 'PSA density',
                                            'MRI EPE (naciek poza torebke)': 'MRI EPE',
                                            'MRI SVI (pecherzyki)': 'MRI SVI',
                                            'MRI SIZE INDEX LESION': 'MRI lesion size',
                                            'ilość\xa0+ wycinków P': 'wycinki P',
                                            'Ilość + wycinków L': 'wycinki L',
                                            'SM +': 'SM+',
                                            'N +': 'N+',
                                            'EPE L RP': 'EPE RP L',
                                            'EPE P RP': 'EPE RP P',
                                            'MRI SIZE': 'MRI lesion size'})
    
    # STUMP in 'Bx ISUP Grade P' column
    first_data.drop(first_data[first_data['Bx ISUP Grade P'] == 'STUMP'].index, inplace=True)

    # empty row
    first_data.drop(41, inplace=True)
    # merge biopsy 
    first_data[['wycinki P +', 'wycinki P -', 'wycinki L +', 'wycinki L -', 'wycinki +', 'wycinki -']] = first_data.apply(lambda row: pd.Series(first_cores(row, 'wycinki P', 'wycinki L')), axis=1)
    first_data['Bx ISUP Grade'] = first_data.apply(lambda row: row[['Bx ISUP Grade P', 'Bx ISUP Grade L']].max(), axis=1)
    first_data['%'] = first_data.apply(lambda row: pd.Series(first_percent(row)), axis=1)

    # calculate PSA density
    first_data['PSA density'] = first_data['PSA'] / first_data['MRI vol']


    return first_data
```
