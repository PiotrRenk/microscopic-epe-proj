# TabPFN

```python
import pandas as pd
from sklearn.metrics import roc_auc_score
from tabpfn import TabPFNClassifier
from unified_functions import *

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../materials/processed_df_new.csv')
df.head()
```

```python
numerical_cols = ['wiek', 'PSA', 'PSAdensity', 'MRI vol', 'MRI SIZE']
categorical_cols = ['MRI Pirads', 'MRI EPE', 'MRI EPE L', 'MRI EPE P', 'MRI SVI', 'MRI SVI L', 'MRI SVI P','Bx ISUP Grade P', 'Bx ISUP Grade L', 'Bx ISUP Grade']
```

```python
X = df[numerical_cols + categorical_cols]
```

```python
def test_tabpfn(target_column):
    y = df[target_column]
    clf = TabPFNClassifier(random_state=42, model_path='../.venv/tabpfn-v2.5-classifier-v2.5_default.ckpt', ignore_pretraining_limits=True)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42, stratify=y)

    clf.fit(X_train, y_train)
    prediction_probabilities = clf.predict_proba(X_test)

    print(f"ROC AUC for {target_column}:", roc_auc_score(y_test, prediction_probabilities[:, 1]))
```

```python
test_tabpfn('EPE RP')
test_tabpfn('EPE micro')
test_tabpfn('EPE macro')
test_tabpfn('SVI RP')
test_tabpfn('N +')
```
